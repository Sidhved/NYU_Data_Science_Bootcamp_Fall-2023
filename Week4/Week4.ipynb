# Q1
- Create a line plot of ZN and INDUS in the housing data.\
    a. For ZN, use a solid green line. For INDUS, use a blue dashed line.\
    b. Change the figure size to a width of 12 and height of 8.\
    c. Change the style sheet to something you find [Link](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html).

```python
df = pd.read_csv("data/boston_housing_data.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(df['ZN'], color='green', label='ZN')
plt.plot(df['INDUS'], color='blue', linestyle='--', label='INDUS')

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of ZN and INDUS in Housing Data')

plt.legend()
```

# Q2
Create a bar chart using col1 and col2 of dummy data.\
    a. Give the plot a large title of your choosing.\
    b. Move the legend to the lower-left corner.\
    c. Do the same thing but with horizontal bars.\
    d. Move the legend to the upper-right corner.

```python
np.random.seed(42)
dummy_data = pd.DataFrame({
    'col1': np.random.randint(1, 10, size=10),
    'col2': np.random.randint(1, 10, size=10)
})
```

```python
plt.figure(figsize=(10, 6))
plt.title("Bar Chart of Col1 and Col2")

plt.bar(dummy_data.index, dummy_data['col1'], label='Col1', color='blue', alpha=0.7)
plt.bar(dummy_data.index, dummy_data['col2'], label='Col2', color='orange', alpha=0.7)

plt.legend(loc='lower left')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Horizontal Bar Chart of Col1 and Col2")

plt.barh(dummy_data.index, dummy_data['col1'], label='Col1', color='blue', alpha=0.7)
plt.barh(dummy_data.index, dummy_data['col2'], label='Col2', color='orange', alpha=0.7)

plt.legend(loc='upper right')

plt.show()
```

# Q3
Create a histogram with pandas for using MEDV in the housing data.\

# Q4
Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
np.random.seed(42)
x_data = np.random.rand(100)
y_data = x_data + np.random.normal(0, 0.1, 100)

plt.scatter(x_data, y_data, color='blue', alpha=0.7)

plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('Scatter Plot with Positive Correlation')

plt.show()
```

# Q5 
Now, create a scatter plot of two heatmap entries that appear to have negative correlation.

```python
np.random.seed(42)
x_data = np.random.rand(100)
y_data = -x_data + np.random.normal(0, 0.1, 100)

plt.scatter(x_data, y_data, color='red', alpha=0.7)

plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('Scatter Plot with Negative Correlation')

plt.show()
```
